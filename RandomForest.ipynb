### Data Cleaning

### Here,I have bserved that the dataset has columns "Unnamed: 32" and "id", which provide no value. Hence, they must be dropped

### Exploratary Data analysis

#### The column **diagnosis** is the target feature, so first lets check out the distribution of the target feature

```python
# Plotting the countplot to see the count of unique values

plt.figure(figsize = (7, 5))
plt.title("The Count of Malignant and Benign")
plt.xlabel("count")
plt.ylabel("diagnosis")
sns.countplot(df.diagnosis, data = df)
plt.show()

df.diagnosis.value_counts()
```

#### The dataset has **357 Benign** values and **212 Malignant** values. This doesn't look like the case of class imbalance, so it'll be safe to use this data to train the model without over/under sampling it

### Checking the Skewness of the features

```python
# Skewness check

df.skew(axis = 0, skipna = True)
```

```python
# Checking features for high skewness

df['diagnosis'] = df['diagnosis'].map({'B':0, 'M':1})
threshold = 1.5
skewed = [feature for feature in df.columns if abs(df[feature].skew()) > threshold]

skewed
```

### It can be observed that the above features as skewed, now checking it via plot

```python
# Checking distribution

plt.figure(figsize = (24, 72))
for i in range(len(skewed)):
    plt.subplot(12, 3, i+1)
    sns.distplot(df[skewed[i]], kde=True)
plt.show()
```

#### It can be seen from the above plot that all these features are left-skewed

#### Let's have a look at feature importance by calculating the feature correlation and plotting the correlation matrix

```python
# Checking the correlation 

corrmat = df.corr() # Correlation Matrix

plt.figure(figsize=(9, 9))

layout = np.zeros_like(corrmat, dtype=np.bool)
layout[np.triu_indices_from(layout)] = True

sns.heatmap(corrmat, mask=layout, vmin=-1.0, vmax=1.0, center=0, cmap = 'RdYlGn',
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

###  checking  which features are most correlated to each other

```python
# Extracting features which have maximum correlation

max_corr_features = []    # Features with maximum correlation

for f1, itr1 in enumerate(corrmat.index):
    for f2, itr2 in enumerate(corrmat.columns):
        if f2 <= f1: continue
        if abs(corrmat.loc[itr1, itr2]) > 0.90:
            print(f' "{itr1}" and "{itr2}" correlation is "{corrmat.loc[itr1, itr2]}" ')
            # print((itr1, itr2), "=> ", round(corrmat.loc[itr1, itr2], 2))
            max_corr_features.append((itr1, itr2))
```

#### checking the scatter plot of highly correlated features 

```python
# Plotting the highly correlated features

plt.figure(figsize = (25, 80))
for feature in range(len(max_corr_features)):
    res = max_corr_features[feature]
    plt.subplot(21, 7, feature + 1)
    sns.scatterplot(x = res[0], y = res[1], hue = "diagnosis", data = df)
plt.show()
```

### It can be observed that all the highly correlated features follow a linear relationship

## Feature Engineering

```python
# Checking head once again

df.head()
```

```python
# Defining Dependent and Independent features

X = df.drop(columns = 'diagnosis')
y = df['diagnosis']
```

```python
# Splitting the data between train and test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Model: Bagging Classifier

from sklearn.ensemble import BaggingClassifier
from sklearn import tree

bagging = BaggingClassifier(tree.DecisionTreeClassifier(random_state=1))
bagging.fit(X_train, y_train)
bagging_score = bagging.score(X_test, y_test)

print(f"The accuracy of Bagging Classifier is {bagging_score * 100 : .3f} % ")
```

```python
# Checking metrics :  Bagging Classifier

from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, plot_confusion_matrix, roc_auc_score, classification_report

y_pred_bagging = bagging.predict(X_test)
roc_score_bagging = roc_auc_score(y_test, y_pred_bagging)
print(f"ROC-AUC Score for Bagging Classifier is {roc_score_bagging} ")
```

```python
# Confusion Matrix : Bagging Classifier

plot_confusion_matrix(bagging, X_test, y_test)
```

```python
# Classification Report : Random Forest

print(classification_report(y_test, y_pred_bagging))
```

```python
# Checking Model Parameters :  Bagging Classifier

bagging.get_params()
```

```python
# N-fold cross validation : Bagging Classifier

from sklearn.model_selection import KFold, cross_val_score

kfold_bagging = KFold(n_splits = 5, random_state = 42, shuffle = True)
score_cv_bagging = cross_val_score(BaggingClassifier(), X_train, y_train, cv = kfold_bagging, scoring = 'accuracy')
print(f"Cross Validation Scores for Bagging Classifier is : {score_cv_bagging}")
print(f"The average of all the scores is {np.mean(score_cv_bagging)}")
print(f"The rough accuracy of the model is {np.mean(score_cv_bagging)*100 : .2f} %")
```

```python
# ROC-Curve : Bagging Classifier

from sklearn.metrics import roc_curve

prob_bagging = bagging.predict_proba(X_test)
fp_bagging, tp_bagging, t1 = roc_curve(y_test, prob_bagging[:, 1], pos_label = 1)

# Plot ROC-Curve

plt.figure(figsize=(6,6))
plt.plot(fp_bagging, tp_bagging, linestyle='--', color='orange', label='Bagging Classifier')
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend(loc='best')
plt.show()
```

```python
# Model: Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state = 42)
clf.fit(X_train, y_train)
rf_score = clf.score(X_test, y_test)

print(f"The accuracy of Random Forest Classifier is {rf_score * 100 : .3f} % ")
```

```python
# Checking metrics

y_pred_rf = clf.predict(X_test)
roc_rf = roc_auc_score(y_test, y_pred_rf)
print(f"ROC-AUC Score for Bagging Classifier is {roc_rf} ")
```

```python
# Classification Report : Random Forest

print(classification_report(y_test, y_pred_rf))
```

```python
# Confusion Matrix : Random Forest

plot_confusion_matrix(clf, X_test, y_test)
```

```python
# Checking Model Parameters :  Random Forest

clf.get_params()
```

```python
# N-fold cross validation : Random Forest

kfold = KFold(n_splits = 5, random_state = 42, shuffle = True)
score_cv = cross_val_score(RandomForestClassifier(), X_train, y_train, cv = kfold, scoring = 'accuracy')
print(f"Cross Validation Score for 5 different models is : {score_cv}")
print(f"The average of all the scores is {np.mean(score_cv)}")
print(f"The rough accuracy of the model is {np.mean(score_cv)*100 : .2f} %")
```

```python
# ROC-Curve : Random Forest

prob_rf = clf.predict_proba(X_test)
fp_rf, tp_rf, t2 = roc_curve(y_test, prob_rf[:, 1], pos_label = 1)

# Plot ROC-Curve

plt.figure(figsize=(6,6))
plt.plot(fp_rf, tp_rf, linestyle='--', color='green', label='Random Forest')
# plt.plot(fpr2, tpr2, linestyle='--',color='green', label='KNN')
# plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend(loc='best')
plt.show()
```

```python
# ROC-Curve Comparision : Bagging Classifier v/s Random Forest

plt.figure(figsize=(6,6))
plt.plot(fp_bagging, tp_bagging, linestyle='--', color='orange', label='Bagging Classifier')
plt.plot(fp_rf, tp_rf, linestyle='--', color='green', label='Random Forest')
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.legend(loc='best')
plt.show()
```
